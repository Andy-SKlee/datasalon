## 6.1.2 서울열린데이터광장의 OPEN API를 활용한 공공데이터 수집

```python
# 예제 6-13  라이브러리 임포트
import requests
import pandas as pd
```

```python
# 예제 6-14  자주 사용하는 변수를 미리 선언
# 등록된 서울열린데이터광장 API 호출 키를 변수에 입력해주세요
SEOUL_API_AUTH_KEY = '76744d4651726f6c3132376e6*******'
```

```python
# 예제 6-15 API 호출을 위한 URL 변수 선언
url = 'http://openapi.seoul.go.kr:8088/{}/json/SdeTlSccoSigW/1/25'.format(SEOUL_API_AUTH_KEY)
print(url)
```

```python
# 예제 6-16 서울열린데이터광장 OPEN API 호출
result_dict = requests.get(url).json()
result_dict
```

```python
# 예제 6-17  시군구 목록 결과 딕셔너리 데이터의 값 가져오기
result_data = result_dict['SdeTlSccoSigW']

print(result_data['list_total_count'])
print(result_data['RESULT'])
print(result_data['row'][0])
```

```python
# 예제 6-18 결과 데이터를 pandas의 데이터프레임으로 변환
data_list = result_data['row']
sample_df = pd.DataFrame(data_list)
sample_df.head()
```

```python
# 예제 6-19  서울열린데이터광장 OPEN API 호출을 위한 공통 함수
# 서울열린데이터광장 API 호출 키
# SEOUL_API_AUTH_KEY = '734*************************'
# 서울열린데이터광장 OPEN API호출 공통 함수
def seoul_open_api_data(url, service):
    data_list = None
    try:
        result_dict = requests.get(url).json()
        result_data = result_dict[service]
        code = result_data['RESULT']['CODE']
        if code == 'INFO-000':
            data_list = result_data['row']
    except:
        pass
    return data_list
```

```python
# 예제 6-20 공통 함수를 이용해 서울시 행정구역 시군구 OPEN API 호출하기
sgg_url = 'http://openapi.seoul.go.kr:8088/{}/json/SdeTlSccoSigW/1/25'.format(SEOUL_API_AUTH_KEY)
sgg_data_list = seoul_open_api_data(url, 'SdeTlSccoSigW')
sgg_data_list[0]
```

```python
# 예제 6-21  결과 데이터를 pandas의 데이터프레임으로 변환
columns = ['SIG_CD','SIG_KOR_NM','LAT','LNG']
sgg_df = pd.DataFrame(data=sgg_data_list, columns=columns)
sgg_df.head()
```

```python
# 예제 6-22  데이터프레임의 칼럼명 변경
sgg_df.columns = ['시군구코드','시군구명','위도','경도']
sgg_df.head()
```

```python
# 예제 6-23  데이터프레임의 요약 정보 확인
sgg_df.info()
```

```python
# 예제 6-24  엑셀로 저장
sgg_df.to_excel('./files/seoul_sgg_list.xlsx', index=False)
```
